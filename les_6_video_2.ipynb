```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
train = pd.read_csv('titanic/train.csv')
train.head(10)
```

```python
train = train.set_index('PassengerId')
```

```python
train
```

```python
train.columns
```

```python
target = 'Survived'
```

```python
y = train[target]
X = train.drop(target, axis=1)
```

```python
X.info()
```

```python
X.drop(['Cabin', 'Name', 'Ticket'], axis=1, inplace=True)
```

```python
X.columns
```

```python
X.info()
```

```python
mean_age = X['Age'].mean()
```

```python
X['Age'] = X['Age'].fillna(mean_age)
X['Age'].unique()
```

```python
X['Embarked'].value_counts()
```

```python
embarked_mode = X['Embarked'].mode()[0]
X['Embarked'] = X['Embarked'].fillna(embarked_mode)
X['Embarked'].value_counts()
```

```python
X['Sex'].value_counts()
```

```python
X['Sex'] = (X['Sex'] == 0).astype(int)
```

```python
X['Sex'].value_counts()
```

```python
X = pd.get_dummies(X)
```

```python
X
```

```python
X['Pclass'] = X['Pclass'].astype(str)
```

```python
X.info()
```

```python
X = pd.get_dummies(X)
```

```python
X.head()
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=1/4, random_state=42)
```

```python
X_train.info()
```

```python
lr = LogisticRegression(solver='lbfgs', max_iter=7600)
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_valid)
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_valid, y_pred)
```

```python
y_pred_train = lr.predict(X_train)
```

```python
accuracy_score(y_train, y_pred_train)
```

```python
y_proba = lr.predict_proba(X_valid)
```

```python
y_proba
```

```python
X_train.to_pickle('X_train.pkl')
X_valid.to_pickle('X_valid.pkl')

y_train.to_pickle('y_train.pkl')
y_valid.to_pickle('y_valid.pkl')
```
